<span id='h101'><h1><a href='#nav_h101'>01.</a> 주제 소개</h1></span><br>
먼저, 제가 선정한 Term-project의 주제가 무엇인지부터 말씀드리겠습니다.

제가 선정한 Term-project 의 주제는 <strong><u>'대중가요로 바라본 2005년~2018년의 대한민국'</u></strong><span id="cont_com_1"><a href="#com_1">[1]</a></span> 입니다.

<span id='h401'><h4><a href='#nav_h401'>(a)</a> 대중가요의 가사</h4></span>

대중가요는 대중들을 향한 음악입니다. 한 특정 시기에 어떤 대중가요 한 곡이 선풍적인 인기를 끌었다면, 아마도 그 곡의 선율이, 박자가, 가사가 그 당시의 대중들의 마음을 크게 자극하였다는 것으로도 볼 수 있을 것입니다.
<br><br>
<div style="width: 40%; background-color: white; box-shadow: 0 4px 8px 0 rgba(0, 0, 0, 0.2), 0 6px 20px 0 rgba(0, 0, 0, 0.19); margin-bottom: 25px;" >
<img src="./report_resources/pop.png">
</div>
더불어, 당대에 유행을 끈 대중가요들은, 시대와 유행에 따라 그 여러가지 특징들이 항상 바뀌어 왔습니다. 

이에 대한 가장 직관적인 예로는, '음악의 장르'를 들 수 있습니다. 2000년대 중반 대한민국 대중가요의 인기 장르는 발라드였습니다. 반면 2010년대 초반에는, 댄스곡들이 대한민국 대중가요의 인기 장르였습니다. 또한 더 나아가 최근 대한민국 대중가요의 인기 장르는 힙합이라고 볼 수 있을 것 같습니다.

저는 이러한 대중가요 속 '유행의 변화'가 혹시 <u>'가사'에도 존재하지 않을까?</u> 라는 생각을 하게 되었습니다. 

이러한 생각이 든 뒤, '그렇다면 각 년도의 히트곡들의 가사를 분석한다면 그 시기 대중들이 추구하던 감성, 원하던 감성이 무엇인지 파악할 수도 있지 않을까?' 라는 호기심을 갖게 되었습니다. 

그리고 그 호기심에 따라, 저는 수 년동안 <u>각 년도마다 대한민국에서 인기를 끌었던 대중가요들의 목록을 수집</u>하기로 결심하였습니다. 그리고 여기에 이어서 곧바로, 각 년도마다 인기를 끌었던 대중가요 곡들의 <u>가사를 분석</u>하여 각 년도마다 대중들이 추구했던 감성과 정서가 무엇인지 찾아보기로 했습니다.

<span id='h402'><h4><a href='#nav_h402'>(b)</a> 인기를 끌었던 가수와 곡</h4></span>

한편, 수 년동안 각 시기마다 인기를 끌었던 대중가요들의 목록 자료를 수집해내는데 성공 해낸다면, 우리는 이 자료로부터 '가사' 데이터의 분석 뿐만 아니라, 또 다른 이런저런 여러 종류의 정보들 또한 분석해볼 수 있을 것입니다. 이에, 과연 또 어떤 정보들을 얻어내볼 수 있을까 고민해봤습니다.

가장 직관적으로, 이 자료들을 분석한다면 우리는 '특정 기간 동안 한국에서 가장 <u>인기가 많았던 곡'</u>들을 알아볼 수 있을 것 같습니다. 또한 마찬가지로 '특정 기간동안 한국에서 가장 <u>인기가 많았던 가수'</u>들도 알아볼 수 있을 것 같습니다. 

이러한 정보들을 도출해낼 수 있다면, '특정 시기의 한국 대중문화에서 큰 축을 차지해왔던 대상들은 도대체 무엇이 있었는지'를 정량적으로 도출해낼 수 있을 것입니다. 대중문화에 관심이 많아왔던 사람들은 이렇게 도출해낸 자료를 보고서 단순히 그 시기의 향수를 떠올리며 감회에 젖을 수 있다는 점만으로도 이러한 자료는 가치가 있을 수도 있습니다. 

하지만 더 의미있어 보이는 것은, 이러한 자료들이 존재한다면, <u>'대중문화에 전혀 관심이 없는 사람들'</u> 마저도, 어느 날 불가피하게 그들이 동시대(또는 특정 시기의 과거)의 대중들은 도대체 어떤 미디어에 푹 빠져있었는지에 대해서 알아야 하는 상황이 왔을 때 그러한 상황을 위 자료를 통해 비교적 간단하게 해결할 수도 있다는 것입니다. 
<br><br><div style="width: 40%; background-color: white; box-shadow: 0 4px 8px 0 rgba(0, 0, 0, 0.2), 0 6px 20px 0 rgba(0, 0, 0, 0.19); margin-bottom: 25px;" >
<img src="./report_resources/jojap.png">
</div>
비슷하게, 어떤 외국인이 출장을 간다던가 하는 이유로 타국의 문화에 빠른 속도로 깊게 스며들어가 하나가 되어야만 하는 상황이 온다면, 그런 상황에서도 이런 자료는 큰 역할을 할 수 있습니다. 이러한 접하기 힘든 특정 외국의 대중문화 정보는 일반적으로 KOTRA의 해외 국가 분석 보고서 중 문화콘텐츠 항목 등을 직접 찾아보는 방식을 통해서만 접할 수 있으며, KOTRA에서 마저도 구체적이고 세세한 자료는 찾을 수 없습니다. 

<span id='h403'><h4><a href='#nav_h403'>(c)</a> 정리</h4></span>

지금까지 서술한 이유들로 인하여, 저는 우리나라에서 수년동안 유행해왔던 대중가요들의 목록을 수집 후, 이를 분석해보기로 했습니다. 그리고 이러한 과정을 짧은 문장으로 정리해보자면 처음에 서술했다시피 <strong><u>'대중가요로 바라본 2005년~2018년의 대한민국'</u></strong> 과 같이 정리될 수 있을 것같습니다.<br>

<span id='h102'><h1><a href='#nav_h102'>02.</a> 분석 대상 및 목표에 대한 구체적인 설명</h1></span><br>

앞서 '01. 주제 소개'에서 제가 할 것들을 크게 3가지로 다음과 같이 정리하였습니다.

<ul>
    <li>1.일정 기간동안 우리나라의 히트곡들의 정보(제목, 가수, 가사)를 전부 획득할 것</li>
    <li>2.획득한 정보를 통해, 파이썬으로 가사를 분석할 것. 이를 통해 당시 대중들이 추구하던 정서를 파악해볼 것</li>
    <li>3.획득한 정보를 통해, 파이썬으로 인기가 특히 많았던 곡과 가수들을 도출할 것</li>
</ul>

이제 여기서는, 위 세가지를 이뤄내기 위해서, 구체적으로 
<ul>
<li>1.'어디서' 정보를 획득할 것인지</li>
<li>2.'어떤 형태로' 정보를 획득할 것인지</li>
<li>3.획득한 특정 형태의 정보를 '어떤 형태로' 가공할 것인지</li>
</ul>

이 세가지에 대해서 <i>구체적으로</i> 설명해보도록 하겠습니다. 즉 여기서는, <u>어떤 출처로부터 우리나라 히트곡 정보를 크롤링 해올 것인지</u>와, 데이터 가공 과정에서 다룰 데이터의 <u>input의 형태</u>와 <u>output의 형태</u>를 여기서 결정해보도록 하는 것입니다.

<span id='h405'><h4><a href='#nav_h405'>(b)</a> Input의 형태</h4></span>

이제 'Input의 형태'를 여기서 결정하고자 합니다.

여기서 제가 'Input의 형태'라는 단어로 지칭하는 것은, 텀프로젝트 중 <u>정보 가공의 과정에서 '재료로 사용할 파일들'의 형태</u>, 즉 데이터 프로세싱 작업에서 <u>'Input으로 집어넣을 파일들'의 형태</u>를 말합니다. 즉, <u>'웹페이지로부터 실제 정보를 어떤 형태, 어떤 규격으로 크롤링 할 것인가'</u>를 여기서 결정하고자 합니다.

일단 제가 얻어내야 되는 실제 정보는 
<ol>
    <li>'2005년~2018년'까지의 모든 주간 차트 정보</li>
    <li>각 차트에 있는 모든 곡들의 가사</li>
    <li>각 차트에 있는 모든 곡들의 제목</li>
    <li>각 차트에 있는 모든 곡들의 가수</li>
</ol>
입니다.

한편, 주간 차트 한개에는 총 100곡(1위~100위)이 들어있습니다. 또한 한 달은 4개의 주간 차트로 이루어져 있습니다(첫째 주, 둘째 주, 셋째 주, 넷째 주). 또한 1년은 12월로 이루어져있습니다. 또한 2005년부터 2018은 총 14년으로 이루어져있습니다. 

즉, 저는 1년 당 100 x 4 x 12 = 4800곡, 총 14년 동안 67200곡의 정보를 받아오고 또 분석해야 합니다. 이러한 상황에서, 저는 <strong>'2종류의 텍스트 파일 양식'</strong>을 통해서 벅스부터 정보를 크롤링 하려고 합니다. 이 중 첫번째 종류의 텍스트 파일은 '가사 정보'를 받아오기 위한 용도의 텍스트 파일입니다. 그리고 이 첫번째 종류의 텍스트 파일의 양식은 다음 이미지와 같습니다.<br><br>

보시다시피 벅스 주간 차트 웹페이지 URL 주소는, 'https://music.bugs.co.kr/chart/track/week/total?chartdate=' 와 같은 문자열의 마지막에 'yyyymmdd' 형태로 8자리 정수 하나를 넣으면, 그 정수에 해당하는 날짜가 '포함되어 있는 주(week)'의 주간차트 웹페이지로 이동시켜주는 URL 주소가 됩니다. 위 이미지에 나와있는 2018년 6월을 기준으로 구체적인 예를 들어보면, URL 주소의 마지막에 20180601 이 들어간 웹페이지, 20180602가 들어간 웹페이지, 20180603이 들어간 웹페이지는 모두 '동일한 하나의 웹페이지', 즉 2018년 6월의 '첫째 주'의 '주간 음원 차트 웹페이지'로 이동시켜주는 URL 주소인 것입니다. <br>

이 정보를 인지하고서 이제 달력을 관찰해보면 다음과 같은 사실을 알 수 있습니다. 모든 '하나의 달(月)'에서는, 각각 '1일(日)', '8일(日)', '15일(日)', '22일(日)', '29일(日)', 이 5개의 일(日)이 포함된 주(week)의 주간차트는 무조건 서로 다른 주간차트일 수 밖에 없습니다. 다시 말해서 '하나의 주간차트'에는 'x월 1일'과 'x월8일'이 절대 동시에 포함될 수 없다는 것입니다. 또한, '29일(日)이 포함된 주'는, 일반적으로 '다음 달의 1일(日)이 포함된 주'와 '동일'하다는 것도 알 수 있습니다.(단, 예외적인 경우가 있습니다. 2018년 7월이 대표적인 예입니다. 7월29일이 포함된 주의 주간차트와, 8월1일이 포함된 주의 주간차트는 서로 '다른 주간 차트' 입니다. 그러나 이런 경우는 매우 소수이기 때문에 경우를 나누는 과정에서 무시하기로 했습니다.)

그렇기에, '2005년 1월'부터 '2018년 12월'까지 모든 달들에 대해서, '1일', '8일', '15일', '22일', 이 4개의 일(日)들로 이루어진 'yyyymmdd' 꼴의 8자리 정수를 파이썬으로 만들고, 이를 벅스 주간 차트 URL 주소의 정수 앞부분인 'https://music.bugs.co.kr/chart/track/week/total?chartdate=' 뒤에 대입하면, 벅스에 존재하는 2005년부터 2018년까지의 '모든 주간차트들의 URL주소'를 생성할 수 있게 됩니다.

이 과정을 실제로 파이썬을 통해 실행하는 것은 다음과 같이 간단히 할 수 있습니다.

```python
mmdd_List = []
for mm in range(1,13):
    for dd in ["01", "08", "15", "22"]:
        #1월~9월인 경우에는 mm에 01,02,03...09가 들어가야 합니다.
        #10월,11월,12월인 경우에는 그냥 mm에 각 달 숫자가 들어가면 됩니다.
        #그렇기에 조건문을 통해 조건을 나눠줍니다.
        if mm<10:
            mmdd_List.append("0"+str(mm)+dd) 
        else:
            mmdd_List.append(str(mm)+dd)
            
yymmdd_List = []
for yy in range(2005, 2019):
    for mmdd in mmdd_List:
        yymmdd_List.append(str(yy)+mmdd)
        
url_List = []
for i in yymmdd_List:
    url_List.append('https://music.bugs.co.kr/chart/track/week/total?chartdate='+i)

#url_List 라는 list type 변수 안에,
#2005년부터 2018년까지의 '모든 주간 음원 차트의 URL주소들'이,
#string type으로 담겨있게 됩니다.


#테스트 코드
print(str(len(url_List))+'개의 url 주소를 만들어냈습니다.')

for i in range(5):
    print(url_List[i])

print('...')

for i in range(5):
    print(url_List[len(url_List)-1-i])
```

위 과정을 통해서 '원하는 모든 웹페이지들의 URL 주소'를 생성해냈습니다.<br>

그러면 이제, '어떤 URL 주소'가 주어졌을 때, '그 URL 주소가 가르키는 웹페이지'의 'html코드' 를 받아오는 과정을 정리해보겠습니다. 이 과정은 Requests 라는 외부 라이브러리를 이용하면 매우 쉽게 할 수 있습니다.<br>(Requests 라이브러리의 공식 Documentation: https://2.python-requests.org/en/master/)

직접 아래 예시를 통해서, URL 주소가 "https://music.bugs.co.kr/chart/track/week/total?chartdate=20190101" 인 주간 차트 웹페이지, 즉 <u>'2019년 1월 1일' 속해있는 주간 음원 차트 웹페이지</u>의 html 코드를 받아보겠습니다. 각 코드에 대한 설명은 코드 내의 주석을 통해 즉각즉각 서술하하겠습니다.

<font color="red">주의: 아래 코드는 반드시 requests 모듈이 깔려있어야만 실행 됩니다</font>

```python
import requests 

inputted_url_String = "https://music.bugs.co.kr/chart/track/week/total?chartdate=20190101"
getted_Object = requests.get(inputted_url_String)
#requests 모듈의 get() 이라는 함수를 사용했습니다.
#get() 함수의 입력인자는 string type data 입니다. 
#그리고 그 string type data는, 다름이 아닌 어떤 웹페이지의 "URL 주소" 입니다.

#반면, get() 함수가 return 하는 data는 어떤 "object" 입니다.
#위 코드에서는 그 data를 'getted_Object'라는 명칭의 변수에 넣은 것을 확인 할 수 있습니다.
#print(type(getted_Object)) 를 통해 이 변수의 type을 확인 가능합니다.

#이 object는 requests모듈만의 고유한 class(설계도)를 통해 만들어진 object(생산품)입니다.
#다시 말해, 웹파이썬 강의에서 배웠듯이, requets 모듈만의 '고유한 새로운 자료형'이라고 생각할 수 있습니다.
#고로, 이 '자료형'만의 새로운 메서드, 어트리뷰트들이 존재합니다.

#이 중 프로젝트에 실제로 필요한 메서드, 어트리뷰트는 극히 일부일 뿐입니다.
#바로 이 아래에서, '.text' 라는 메서드를 이용합니다.

getted_Html = getted_Object.text

#이 object 안에 있는 text 라는 attribute 를 getted_Html 이라는 이름의 변수에 담았습니다.
#이 attribute의 정체는 뭘까요?

print(type(getted_Html))
print(len(getted_Html))


#바로, 자료형은 string 이고, 그 내용은 '실제 웹페이지의 html 코드' 인 data 입니다.
#이 메서드를 통해서, 마침내 원래의 목적이었던 '어떤 웹페이지의 URL 이 string type 으로 주어졌을 때, 그 웹페이지의 html을 string type 으로 받아오는 것'을 완수할 수 있게 된 것입니다..
```

위 코드를 통해서 모든 차트 웹페이지의 html 코드들을, string type으로 받아와서 파일로 저장해뒀습니다. 받아온 string type의 html코드는, 쥬피터 노트북에서 print(getted_Html)을 시행하면 확인하실 수 있습니다. html 코드의 길이가 꽤 길기에, 본 보고서에서는 Slicing 을 이용해 앞부분 일부만 출력하도록 하겠습니다.

```python
print(getted_Html[:1000], "\n...")
```

이제 받아온 html 자료를 파이썬에서 분석 하고자 합니다. 좀 더 구체적으로 말해보면, 이 html 코드에서 어떤 태그가 어디에 있는지, 그 태그에는 어떤 text가 담겨있는지, 등등의 정보를 파이썬을 통해서 추출해내고자 하는 것입니다. 이러한 작업을 일반적으로 'html 코드를 Parsing 한다, 파싱한다' 라는 문장으로 표현하기도 합니다. 파이썬에서는 이 'html 코드의 Parsing' 이라는 작업을 BeautifulSoup 이라는 외부 라이브러리 통해 할 수 있습니다.<br>(BeautifulSoup github:https://github.com/waylan/beautifulsoup)<br>
(BeautifulSoup Official Documentation:https://www.crummy.com/software/BeautifulSoup/bs4/doc/#)<br>

먼저 <u>'가사의 명사 빈도를 분석하여 WordCloud를 만들어보기'</u> 를 해보겠습니다. 이를 위해서는, 일단 string type의 '가사 정보' 에서 '명사'들을 추출해내야 합니다. 이를 위해서는 자연어 처리 기술이 필요합니다. 파이썬에서는 고맙게도 이미 존재하는 'KoNLPy'<span id="cont_com_3"><a href="#com_3">[3]</a></span>라는 라이브러리를 통해 이런 작업이 가능합니다. 

이제, 아래 예시를 통해서 직접 어떤 'string type의 정보'로부터 '형태소'를 추출고나서 이를 'list type의 정보'로 묶어보겠습니다.(단, 실제 프로젝트에서는 전체 가사에서 '형태소'가 아닌 '명사'를 추출하였습니다. 형태소를 전부 추출할 경우 데이터의 규모가 너무 커졌기 때문입니다.)<br>

이 때 예시로 사용할 'string type'의 정보는, 위 03장에서 이미 예시로 받아온 '20190101이 속해있는 주간 음원 차트 웹페이지의 1위곡'인, '청하 - 12시'의 가사로 하겠습니다. 이 정보는 이미 03장을 거쳐서 본 .ipynb 파일의 파이썬에서 'lyrics_String' 이라는 이름의 변수안에 담겨있습니다.

<font color="red">주의: 아래 코드는 반드시 KoNLPy 모듈이 깔려있어야만 실행 됩니다</font><span id="cont_com_4"><a href="#com_4">[4]</a></span>

```python
from konlpy.tag import Okt
#Okt는 konlypy에서 제공하는 '파이썬 클래스'입니다.
#konlpy 에서 제공하는 여러 한글 자연어 분석 도구들 중 한 가지입니다.
#속도가 제일 빨라 사용하였습니다.


NLpro = Okt() #object를 만들었습니다.
morphs_List = NLpro.morphs(lyrics_String)
#konlpy의 공식 매뉴얼에 기본적인 사용법과 문법이 나와있습니다.
#.morphs() 함수를 이용하면 string 을 input으로 받고, input string 을 형태소로 나눈 후 list로 묶습니다. 그리고 그 묶은 list를 output으로 반환합니다.
#이 예시 코드가 아닌, 실제 프로젝트의 파이썬 코드에서는 .nouns() 함수를 사용했습니다. morphs() 함수와 사용법이 완전히 동일합니다.
print(morphs_List)
print(type(morphs_List))
```

출력 결과에서 보이다시피, 의미가 없는 한글자 짜리 형태소와 '\r\n' 같은 이스케이프 구문들까지 전부 추출된 것을 확인 할 수 있습니다. 이런 의미없는 elements들을 제거한 새로운 리스트를 만들어보도록 하겠습니다.

```python
new_List = []
for i in morphs_List:
    if (len(i)>1) and (i != "\r\n"):
        new_List.append(i)

print(new_List)
```

이제 원하는 형태의 추출물, 원하던 내용의 list가 만들어졌습니다. 여기서, 이제 <u>어떤 단어가 몇 번 나왔는지</u>를 조사해야 합니다. 즉, 단어가 나온 빈도를 계산해야 합니다. 파이썬에서는 고맙게도 collections 라는 내장 모듈의 Counter 라는 클래스를 통해 이러한 작업을 지원해줍니다. 

```python
from collections import Counter

word_Counts = Counter(new_List)
print(word_Counts)
print(type(word_Counts))

print("\n\n자료형 변환 후:\n")
wordcounts_Dict = dict(word_Counts) 
#우리에게 익숙한, 파이썬의 기본 자료형인 dictionary type으로 바꿔줍니다.
#고맙게도 Counter 클래스로 만든 object는 기본적인 함수인 dict() 함수에 input으로 넣어주기만 해도 output으로 dictionary 형태의 정보가 나옵니다.
print(wordcounts_Dict)
print(type(wordcounts_Dict))
```

빈도 계산이 끝났으니 드디어 이를 통해 'WordCloud'를 만들어 볼 차례입니다. 파이썬에서는 간단하게 'wordcloud' 라는 외부 모듈을 통해서 이 작업이 가능합니다. 이 모듈의 공식 웹 링크는 다음과 같습니다.<br>

<span id='h408'><h4><a href='#nav_h408'>(b)</a> 인기 곡, 가수 분석 과정</h4>

이제 두번째 단계, 즉 2005년부터 2018년까지, <u>차트에 가장 자주 올라왔던 곡과 가수들을 찾아내 bar graph 로 표현</u>하는 것을 해보겠습니다. 이 때 03장의 과정에서 받아온 14개의 'title_and_artistxxxx.txt' 파일들 중 한개인 'title_and_artist2005.txt' 파일을 써보도록 하겠습니다. 제목 그대로 2005년도의 모든 주간 차트 웹페이지에 올라왔던 곡제목과 가수 정보를 담은 텍스트 파일입니다.

먼저 해당 파일의 텍스트 내용이 어떤 식으로 되어있나 그 구조를 먼저 살펴보겠습니다.

```python
text_File = open("./report_resources/title_and_artist2005.txt", "r", encoding="utf-8")
content_String = text_File.read()
text_File.close()
print(content_String[:300])
#텍스트 파일의 크기가 굉장히 크니 300글자 까지만 출력하도록 하겠습니다.
print("...")
```

텍스트 구조를 보니, '가장 여러번 올라왔던 곡'을 찾을 때는 그냥 본 텍스트 파일을 한 줄 한 줄 읽어서 하나의 요소로 다루면 될 것 같습니다.</br>

반면, '가장 여러번 올라왔던 가수'를 찾을 때는, 한 줄을 '/' 문자를 기준으로 나누고서, 이 중 오른쪽에 있는 요소를 저장해야 될 것 같습니다.<br>

이렇게 하여 이 텍스트 파일로부터, '곡들 목록'과 '가수들 목록'을 list 형태의 변수로 저장해보도록 하겠습니다. 그리고 앞에서 wordcloud를 그릴 때 사용했던 collections.Counter 클래스를 이용하여, 아예 '곡들이 차트에 올라왔던 횟수'와 '가수가 차트에 올라왔던 횟수' 까지 전부 다 계산해보도록 하겠습니다.

```python
tracks_List = content_String.split("\n")
artists_List = []

for i in tracks_List:
    try:
        artists_List.append(i.split("/")[1])
    except:
        pass

tracks_Count = Counter(tracks_List) 
#wordcloud를 그릴 때 이미 collections 모듈안의 Counter 클래스를 import 했었습니다.
#그래서 이 코드에서는 Counter 클래스를 따로 import를 해주지 않아도 오류 없이 위 내용이 잘 실행됩니다.
artists_Count = Counter(artists_List)

mostCommon_tracks = tracks_Count.most_common(20)
#.most_common은 Counter object의 고유한 메서드입니다. 제일 횟수가 많은 요소 n개를 반환하여 list로 묶어주는 메서드입니다.
mostCommon_artists = artists_Count.most_common(20)

print(mostCommon_tracks,"\n")
print(mostCommon_artists)
```

이제, 앞에서 wordcloud 결과물을 화면에 뿌려줄 때도 사용했었던 matplotlib라는 외부 라이브러리를 이용하여, 저 통계 결과를 '막대 그래프'로 표현해보도록 하겠습니다. 이 matplotlib 라는 라이브러리의 원래 용도는 '파이썬에서 2D 그래프를 그리는 것' 입니다.<br>

matplotlib를 통해 '가로 막대 그래프'를 그려내는 방법은 아래와 같습니다.

<font color="red">주의: 아래 코드는 반드시 matplotlib 모듈이 깔려있어야만 실행 됩니다</font>

```python
list(dict(mostCommon_tracks).keys())
list(dict(mostCommon_tracks).values())
from matplotlib import font_manager, rc

font_name = font_manager.FontProperties(fname="./report_resources/malgun.ttf").get_name()
rc('font', family=font_name)

print("차트에 제일 여러번 올라왔던 곡 20개")
pyplot.barh(list(dict(mostCommon_tracks).keys())[::-1], list(dict(mostCommon_tracks).values())[::-1])
pyplot.show()

print("차트에 제일 여러번 올라왔던 가수 20명")
pyplot.barh(list(dict(mostCommon_artists).keys())[::-1], list(dict(mostCommon_artists).values())[::-1])
pyplot.show()
```

본 프로젝트에서는 차트에 올라온 각 곡들에 대해서, 오로지 '차트에 올라온 횟수' 만 고려되었습니다. 다시 말해 '차트에서의 순위'는 전혀 고려하지 않았습니다. 예컨대 1위를 5번 한 곡과, 100위를 6번 한 곡이 있다고 할 때, 후자의 경우가 더 '인기가 많은 곡'으로 결정되는 방식으로 통계가 이루어졌습니다. 이 부분은 영 합리적이지 않아 보입니다. 

만약 '차트에서의 순위' 또한 곡과 가수의 인기도의 판단 기준으로써 '차트에 올라온 횟수'와 함게 잘 조합할 수 있는 방법이 있다면, 차이가 크지는 않더라도 약간이라도 다른 새로운 결과가 나올 것 같습니다.

더불어 각 곡들에 대한 '음원 재생 횟수' 정보는 벅스 웹싸이트에서 확인할 수 있는 방법을 못 찾았습니다. 이 부분 또한 아쉬웠습니다. 만약 해당 정보에 대한 크롤링도 성공했었다면, 어떤 가수가 특정 기간동안 음원 재생으로 얼마의 수익을 얻었는지, 가수들이 음원 재생 평균 수입은 어떻게 되는지 등등의 정보도 알 수 있었을 것 같습니다.<br><br>
